```python
# merge commons
inner_join_df = pd.merge(leff_df[['IsomericSMILES', 'descriptorsL']], goods_df[['IsomericSMILES', 'descriptorsG']], how='inner', on='IsomericSMILES')
```

```python
def merge_desc(desc_combined):
    desc_list = desc_combined.split(';')
    new_desc_set = set()
    for desc in desc_list:
        if desc == 'odorless':
            continue
        new_desc_set.update([desc])
    if len(new_desc_set) == 0:
        new_desc_set.update(['odorless'])
    return ';'.join(new_desc_set)
```

```python
merge_desc('odorless;phenolic;nutty;nutty')
```

```python
merge_desc('odorless;odorless')
```

```python
inner_join_df['merged_descriptors'] = inner_join_df.apply(lambda x: merge_desc(x['descriptorsL']+';'+x['descriptorsG']), axis=1)
```

handle stereo isomers

```python
from rdkit import Chem

def canonical_smiles(smiles):
    return Chem.MolToSmiles(Chem.MolFromSmiles(smiles), isomericSmiles = True)
```

```python
def remove_stereo(smiles):
    smiles = smiles.replace('@','')
    smiles = smiles.replace('/','')
    smiles = smiles.replace('\\','')
    return canonical_smiles(smiles)
```

```python
remove_stereo('CC/C=C\CC(C)OC(=O)@C')
```

```python
remove_stereo('CCC/C=C(\C)/C(=O)O')
```

```python
remove_stereo('CC1=CC(=C(O1)C)SC(=O)CC(C)C')
```

```python
combined_dataset_encoded['nonStereoSMILES'] = combined_dataset_encoded['IsomericSMILES'].apply(lambda x: remove_stereo(x))
```

```python
combined_dataset_encoded[combined_dataset_encoded['nonStereoSMILES'].duplicated(keep=False)]
```

```python
dup = combined_dataset_encoded[['nonStereoSMILES', 'descriptors']][combined_dataset_encoded['nonStereoSMILES'].duplicated(keep=False)]
index_set = set()
index_set.update(dup['nonStereoSMILES'].to_list())
for i in index_set:
    print(dup[dup['nonStereoSMILES'] == i])
    print()
```

```python
dup = combined_dataset_encoded[['nonStereoSMILES', 'descriptors']][combined_dataset_encoded['nonStereoSMILES'].duplicated(keep=False)]
index_set = set()
index_set.update(dup['nonStereoSMILES'].to_list())
list_idx_of_duplicates = []
for i in index_set:
    # print(list(dup[dup['CID'] == i].index))
    list_idx_of_duplicates.append(list(dup[dup['nonStereoSMILES'] == i].index))
list_idx_of_duplicates
```

```python
combined_dataset_encoded['descriptors_old'] = combined_dataset_encoded.loc[:, 'descriptors']
combined_dataset_encoded.head()
```

```python
for dup_set in list_idx_of_duplicates:
    desc_set = set()
    desc_string = ';'.join(combined_dataset_encoded['descriptors'][dup_set].to_list())
    for desc in desc_string.split(';'):
        if desc == 'odorless':
            continue
        desc_set.update([desc])
    if len(desc_set) == 0:
        desc_set.update(['odorless'])
    new_desc_string = ';'.join(desc_set)
    combined_dataset_encoded['descriptors'][dup_set] = new_desc_string

combined_dataset_encoded_cleaned = combined_dataset_encoded.drop_duplicates('nonStereoSMILES', keep='first').reset_index(drop=True)
```
